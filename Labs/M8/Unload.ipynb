You can use the UNLOAD command in Redshift to unload data from a table to S3

```redshift
UNLOAD ('SELECT * FROM spectrum_schema.sales_data')
TO 's3://niks-glue-data/output/'
CREDENTIALS 'aws_iam_role=arn:aws:iam::146962103229:role/RedshiftAccessRole'
DELIMITER ','
ADDQUOTES
ALLOWOVERWRITE
PARALLEL OFF;
```

To unload data in JSON format, you can use the following command

```redshift
UNLOAD ('SELECT * FROM spectrum_schema.sales_data')
TO 's3://niks-glue-data/output/'
CREDENTIALS 'aws_iam_role=arn:aws:iam::146962103229:role/RedshiftAccessRole'
ALLOWOVERWRITE
FORMAT AS JSON;
```

Parquet Format

```redshift
UNLOAD ('SELECT * FROM spectrum_schema.sales_data')
TO 's3://niks-glue-data/output/'
CREDENTIALS 'aws_iam_role=arn:aws:iam::146962103229:role/RedshiftAccessRole'
PARQUET
ALLOWOVERWRITE;
```

ORC Format
Not supported

```redshift
UNLOAD ('SELECT * FROM spectrum_schema.sales_data')
TO 's3://niks-glue-data/output/'
CREDENTIALS 'aws_iam_role=arn:aws:iam::146962103229:role/RedshiftAccessRole'
ORC
ALLOWOVERWRITE;
```

Unloading Specific Columns or Filters

```redshift
UNLOAD ('SELECT order_id, customer_id, total_amount FROM spectrum_schema.sales_data WHERE order_date >= ''2023-01-01''')
TO 's3://niks-glue-data/output/'
CREDENTIALS 'aws_iam_role=arn:aws:iam::146962103229:role/RedshiftAccessRole'
DELIMITER ','
ADDQUOTES
ALLOWOVERWRITE;
```
